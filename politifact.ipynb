```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.politifact.com/'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup)
```

```python
all_text = soup.get_text(separator="\n", strip=True)

# Print the extracted text
print(all_text)
```

```python
# Extract specific sections of the page based on your analysis
# This is just an example, and you should customize it based on your needs

# Extract headlines with <h1> tags
headlines = soup.find_all('h1')
for headline in headlines:
    print(f'<h1>{headline.get_text(strip=True)}</h1>')

# Extract paragraphs with <p> tags
paragraphs = soup.find_all('p')
for paragraph in paragraphs:
    print(f'<p>{paragraph.get_text(strip=True)}</p>')

# Extract links with <a> tags
links = soup.find_all('a')
for link in links:
    print(f'<a href="{link.get("href")}">{link.get_text(strip=True)}</a>')
```

```python
# Extract data
headlines = [headline.get_text(strip=True) for headline in soup.find_all('h1')]
paragraphs = [paragraph.get_text(strip=True) for paragraph in soup.find_all('p')]
links = [(link.get('href'), link.get_text(strip=True)) for link in soup.find_all('a')]
```

```python
print("Headlines:", headlines)
print("Paragraphs:", paragraphs)
print("Links:", links)
```

```python
# Create DataFrames
headlines_df = pd.DataFrame({'Headlines': headlines})
paragraphs_df = pd.DataFrame({'Paragraphs': paragraphs})
links_df = pd.DataFrame({'Link': [link[0] for link in links], 'Link Text': [link[1] for link in links]})
```

```python
print(headlines_df)
print(paragraphs_df)
print(links_df)
```

```python
# Save DataFrames to separate files
headlines_df.to_csv('headlines.csv', index=False)
paragraphs_df.to_csv('paragraphs.csv', index=False)
links_df.to_csv('links.csv', index=False)
```

# Extracting videos

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.politifact.com/"  # Replace with the actual URL of your webpage

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Target video elements using a more flexible selector
video_elements = soup.find_all("video[src]")

# Extract video URLs
video_urls = [video['src'] for video in video_elements]

if not video_urls:
    print("No videos found on the webpage.")
else:
    # Print the extracted video URLs
    for video_url in video_urls:
        print(video_url)
```
